### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
df["Attrition"].unique()
```

```python
def zeroone(x):
    if x == "Yes":
        return 1
    elif x == "No":
        return 0
```

```python
df["Attrition"] = df["Attrition"].apply(zeroone)
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf = pd.get_dummies(df,drop_first=True,dtype="int")
newdf.head()
```

#### Check the shape of our new dataset

```python
newdf.shape
```

#### Print unique values in our new dataframe

```python
for i in newdf.columns:
    num_unique = newdf[i].unique()
    print(f"Number of unique elements in {i} are : ",num_unique)
```

#### Split the columns into input and target variables (X and y)

```python
x = newdf.drop(columns=["Attrition"])
y = newdf["Attrition"]
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xcolumns = x.columns
```

```python
x = scaler.fit_transform(x)
```

```python
x = pd.DataFrame(x,columns=xcolumns)
```

```python
x.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accuracy score of our model is : ",accuracy_score(y_test,y_pred))
print("Confusion matrix is : \n",confusion_matrix(y_test,y_pred))
print("Classification report is: \n",classification_report(y_test,y_pred))
cv_score = cross_val_score(LogisticRegression(),x,y,cv=3).max()
print("Cross validation Score is: ",cv_score)
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")
plt.show()
```

### Random Forest Classifier
** Choose the best estimator and parameters :GridSearchCV**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params = {'max_depth':[0.5,1, 5, 10],'n_estimators': [16, 32, 50, 100], 'n_jobs': [1,2],'random_state':[1,10,20,42]}
```

```python
gsv = GridSearchCV(RandomForestClassifier(), params, scoring = 'accuracy')
```

```python
print(gsv.fit(x_train,y_train))
```

```python
print("Best estimator is : ",gsv.best_estimator_)
print("Best parameters are : ",gsv.best_params_)
print("Best score is : ",gsv.best_score_)
```

<span style = "color:blue"> Create Random forest model with the best parameters </span>

```python
best_params = gsv.best_params_
rf_model = RandomForestClassifier(max_depth = best_params["max_depth"], n_estimators=best_params["n_estimators"],
                                  n_jobs=best_params["n_jobs"], random_state=best_params["random_state"]
                                 )
```

```python
print(rf_model.fit(x_train,y_train))
```

```python
rf_y_pred = rf_model.predict(x_test)
```

```python
print("Accuracy score of our model is : ",accuracy_score(y_test,rf_y_pred))
print("Confusion matrix is : \n",confusion_matrix(y_test,rf_y_pred))
print("Classification report is: \n",classification_report(y_test,rf_y_pred))
rf_cv_score = cross_val_score(RandomForestClassifier(),x,y,cv=3).max()
print("Cross validation Score is: ",rf_cv_score)
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test,rf_y_pred),annot=True, fmt="d")
plt.show()
```
